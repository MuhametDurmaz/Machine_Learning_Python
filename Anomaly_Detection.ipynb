```python
# Function for determining gaussan distribution
def estimate_gaussian(X): 

    m, n = X.shape

    mu=np.sum(X,axis=0)/m
    var=np.sum((X-mu)**2,axis=0)/m
    return mu, var
```

```python
# Function for determining the optimum probability threshold for anomaly detection
def select_threshold(y_val, p_val): 
    best_epsilon = 0
    best_F1 = 0
    F1 = 0
    
    step_size = (max(p_val) - min(p_val)) / 1000
    
    for epsilon in np.arange(min(p_val), max(p_val), step_size):

        predictions=p_val<epsilon
        tp=np.sum((predictions==y_val) & (predictions==1))
        fn =np.sum((predictions!=y_val) & (predictions==0))
        prec =tp/np.sum(predictions)
        rec= tp/(tp+fn)
        F1= (2*prec*rec)/(prec+rec)

        if F1 > best_F1:
            best_F1 = F1
            best_epsilon = epsilon
        
    return best_epsilon, best_F1
```
